```python
from types import SimpleNamespace

import pandas as pd

# 로컬 CSV에서 데이터 읽기
df = pd.read_csv("diabetes.csv")

# 기존 예제 코드와 동일한 속성 구조로 변환
feature_names = df.columns.drop("target").tolist()
diabetes = SimpleNamespace(
    data=df[feature_names].to_numpy(),
    target=df["target"].to_numpy(),
    feature_names=feature_names,
)

diabetes
```

```python
import numpy as np
import pandas as pd

# 설정값
target_col = "target"
ratio_threshold = 10.0   # 예: 2.0이면 "최대 평균이 최소 평균의 2배 이상"만 출력
eps = 1e-9              # 0으로 나누기 방지

# 현재 노트북에서는 ford가 분석 대상
data = ford

if target_col not in data.columns:
    raise KeyError(f"'{target_col}' 컬럼이 없습니다. 현재 컬럼: {list(data.columns)[:10]} ...")

# 숫자형 변수만 대상으로 (target 제외)
num_cols = [c for c in data.select_dtypes(include=[np.number]).columns if c != target_col]

# target별 평균
mean_by_target = data.groupby(target_col)[num_cols].mean()

rows = []
for col in num_cols:
    s = mean_by_target[col]

    # 절대값 기준 비율(음수/양수 섞여도 비교 가능)
    abs_vals = s.abs()
    min_abs = abs_vals.min()
    max_abs = abs_vals.max()
    ratio = max_abs / max(min_abs, eps)

    if ratio >= ratio_threshold:
        rows.append({
            "variable": col,
            "ratio(max_abs/min_abs)": ratio,
            **{f"mean_target_{t}": v for t, v in s.items()}
        })

# 결과 출력
result = pd.DataFrame(rows).sort_values("ratio(max_abs/min_abs)", ascending=False)

if result.empty:
    print(f"조건(ratio >= {ratio_threshold})에 해당하는 변수가 없습니다.")
else:
    print(result.to_string(index=False))
```
